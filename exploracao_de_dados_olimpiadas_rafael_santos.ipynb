```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("/content/world_olympedia_olympics_game_medal_tally.csv")
pd.options.display.float_format = "{:.0f}".format


summer_filter = df["edition"].str.contains("Summer", na=False)
summer_data = df[summer_filter].groupby("country")["total"].sum()

summer_top50 = summer_data.sort_values(ascending=False).head(50)


plt.figure(figsize=(10, 50))
plt.barh(summer_top50.index, summer_top50.values, color="skyblue")
plt.title("Top 10 - Medalhas Totais em Jogos Olímpicos de Verão")
plt.xlabel("Total de Medalhas")
plt.gca().invert_yaxis()
plt.show()
```

```python
winter_filter = df["edition"].str.contains("Winter", na=False)
winter_data = df[winter_filter].groupby("country")["total"].sum()

winter_top50 = winter_data.sort_values(ascending=False).head(50)

plt.figure(figsize=(10, 50))
plt.barh(winter_top50.index, winter_top50.values, color="salmon")
plt.title("Top 10 - Medalhas Totais em Jogos Olímpicos de Inverno")
plt.xlabel("Total de Medalhas")
plt.gca().invert_yaxis()
plt.show()
```

```python

total_data = df.groupby("country")["total"].sum()

total_top50 = total_data.sort_values(ascending=False).head(50)

plt.figure(figsize=(10, 50))
plt.barh(total_top50.index, total_top50.values, color="lightgreen")
plt.title("Top 10 - Medalhas Totais em Toda a História Olímpica")
plt.xlabel("Total de Medalhas")
plt.gca().invert_yaxis()
plt.show()
```
